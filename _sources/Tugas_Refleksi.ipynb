A. Pengertian Transformasi Geometri

Transformasi geometri adalah proses memindahkan atau mengubah posisi suatu objek pada bidang koordinat tanpa mengubah bentuk dasarnya.

Pada materi ini digunakan dua jenis transformasi:

Translasi (pergeseran)
Refleksi (pencerminan)

B. Translasi (Pergeseran)

Translasi adalah pergeseran suatu titik sejauh tertentu.

Rumus translasi:

x
′
=x−t

Keterangan:

x = koordinat awal
x
′
 = koordinat setelah translasi
t = besar pergeseran

👉 Pada program:

Pergeseran dilakukan ke kiri, sehingga dikurangi
Nilai t berubah tiap langkah: 0, 1, 2, 3

C. Refleksi terhadap Sumbu Y

Refleksi adalah pencerminan suatu titik terhadap suatu garis (di sini sumbu Y).

Rumus refleksi:

x
′′
=−x
′

Keterangan:

x
′
 = hasil translasi
x
′′
 = hasil refleksi

👉 Artinya:

Nilai x dibalik tanda (+ jadi −, − jadi +)
Nilai y tetap

```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# =========================
# OBJEK AWAL
# =========================
obj1 = [(3, 2), (5, 2), (4, 4)]

x_awal = [p[0] for p in obj1]
y_awal = [p[1] for p in obj1]

fig, ax = plt.subplots()

# step jelas (bukan pecahan)
steps = [0, 1, 2, 3]

def update(frame):
    ax.clear()

    geser = steps[frame]  # langsung lompat 1 satuan

    # =========================
    # TRANSLASI
    # =========================
    x1_anim = [x - geser for x in x_awal]
    y1_anim = y_awal

    # =========================
    # REFLEKSI
    # =========================
    x2_anim = [-x for x in x1_anim]
    y2_anim = y1_anim

    # =========================
    # PRINT PROSES
    # =========================
    print(f"\nSTEP {frame}")
    print(f"Geser: {geser}")
    print("Objek 1:", list(zip(x1_anim, y1_anim)))
    print("Objek 2:", list(zip(x2_anim, y2_anim)))

    # =========================
    # GAMBAR
    # =========================
    ax.plot(x1_anim + [x1_anim[0]], y1_anim + [y1_anim[0]],
            marker='o', label='Objek 1 (Translasi)')

    ax.plot(x2_anim + [x2_anim[0]], y2_anim + [y2_anim[0]],
            marker='o', linestyle='--', label='Objek 2 (Refleksi)')

    ax.axhline(0)
    ax.axvline(0)

    ax.set_xlim(-5, 5)
    ax.set_ylim(-2, 5)
    ax.set_title(f"Step {frame} | Geser {geser}")
    ax.legend()
    ax.grid()

ani = FuncAnimation(fig, update, frames=len(steps), interval=800)

HTML(ani.to_jshtml())
```
