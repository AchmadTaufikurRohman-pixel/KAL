```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# =========================
# OBJEK AWAL (KANAN)
# =========================
obj1 = [(3, 2), (5, 2), (4, 4)]

x_awal = [p[0] for p in obj1]
y = [p[1] for p in obj1]

fig, ax = plt.subplots()

def update(frame):
    ax.clear()

    # =========================
    # STEP TRANSLASI
    # =========================
    geser = frame  # 0,1,2,3

    # objek 1 (geser ke kiri)
    x1_anim = [x - geser for x in x_awal]
    y1_anim = y

    # =========================
    # OBJEK 2 = REFLEKSI
    # =========================
    x2_anim = [-x for x in x1_anim]
    y2_anim = y1_anim

    # gambar objek 1
    ax.plot(x1_anim + [x1_anim[0]], y1_anim + [y1_anim[0]],
            marker='o', label='Objek 1 (Translasi)')

    # gambar objek 2 (bayangan)
    ax.plot(x2_anim + [x2_anim[0]], y2_anim + [y2_anim[0]],
            marker='o', linestyle='--', label='Objek 2 (Refleksi)')

    # sumbu
    ax.axhline(0)
    ax.axvline(0)

    ax.set_xlim(-10, 10)
    ax.set_ylim(-2, 5)
    ax.set_title(f"Step {frame}: Translasi + Refleksi")
    ax.legend()
    ax.grid()

# step: 3 → 2 → 1 → 0
ani = FuncAnimation(fig, update, frames=4, interval=800)

HTML(ani.to_jshtml())
```
